```python
main=df.copy()
nl=main.ingredients.str.split(',').explode()
count_inge=nl.value_counts()
top20=count_inge.index[:20]
top10=count_inge.index[:10]

```

```python
f=df.ingredients.str.split(',')

idx_ing=nl.value_counts().index
top50=idx_ing[:50]

```

```python
new_ingredients = []
for ingredients_list in f:
  filtered_ingredients = [ing for ing in ingredients_list if ing in top50]
  new_ingredients.append(','.join(filtered_ingredients))

df['top_50_ingredients'] = new_ingredients
```

```python
df
```

```python

nl = nl.where(nl.isin(top20), other=None)

ko=sorted(nl.value_counts().index)
plt.figure(figsize=(10,5))
sns.countplot(nl)
plt.xticks(rotation=90)
```

```python
def topn(ser,n=5):
  vals=ser.value_counts()
  return ser.where(ser.isin(vals.index[:n]),other=None)
nl.pipe(topn,5).value_counts().plot.barh().legend(bbox_to_anchor=(1,1))
```
